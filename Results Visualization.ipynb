```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bert_features_df = pd.read_csv('results/BERT + features.csv', index_col = 0)
deberta_df = pd.read_csv('results/DeBERTa.csv', index_col = 0)
distilbert_df = pd.read_csv('results/DistilBERT.csv')
gpt2_df = pd.read_csv('results/GPT_2.csv')
gpt2_features_df = pd.read_csv('results/gpt-2 + features.csv', index_col = 0)
```

```python
bert_features_df.rename(index = {'BERT + LR': 'BERT + LR (Features)', 
                                'BERT + RF': 'BERT + RF (Features)',
                                'BERT + Adaboost': 'BERT + Adaboost (Features)',
                                'BERT + KNN': 'BERT + KNN (Features)',
                                'BERT + SVC': 'BERT + SVC (Features)'}, inplace = True)
```

```python
gpt2_features_df.rename(index = {'GPT-2 + LR': 'GPT-2 + LR (Features)', 
                                'GPT-2 + RF': 'GPT-2 + RF (Features)',
                                'GPT-2 + Adaboost': 'GPT-2 + Adaboost (Features)',
                                'GPT-2 + KNN': 'GPT-2 + KNN (Features)',
                                'GPT-2 + SVC': 'GPT-2 + SVC (Features)'}, inplace = True)
```

```python
distilbert_df.rename(columns = {'0': 'Prompt-1', '1': 'Prompt-2',
                               '2': 'Prompt-3', '3': 'Prompt-4',
                               '4': 'Prompt-5', '5': 'Prompt-6',
                               '6': 'Prompt-7', '7': 'Prompt-8'},
                    index = {0: 'distilBERT + LSTM', 1: 'distilBERT + biLSTM',
                            2: 'distilBERT + CNN', 3: 'distilBERT + LR',
                             4: 'distilBERT + RF', 5: 'distilBERT + Adaboost', 
                             6: 'distilBERT + KNN', 7: 'distilBERT + SVC'},
                    inplace = True)
```

```python
gpt2_df.rename(columns = {'0': 'Prompt-1', '1': 'Prompt-2',
                               '2': 'Prompt-3', '3': 'Prompt-4',
                               '4': 'Prompt-5', '5': 'Prompt-6',
                               '6': 'Prompt-7', '7': 'Prompt-8'},
                    index = {0: 'GPT-2 + LSTM', 1: 'GPT-2 + biLSTM',
                            2: 'GPT-2 + CNN', 3: 'GPT-2 + LR',
                             4: 'GPT-2 + RF', 5: 'GPT-2 + Adaboost', 
                             6: 'GPT-2 + KNN', 7: 'GPT-2 + SVC'},
                    inplace = True)
```

```python
gpt2_df
```

```python
distilbert_df
```

```python
df_results_final = pd.concat([bert_features_df, deberta_df, distilbert_df, gpt2_df, gpt2_features_df])
```

```python
df_results_final
```

```python
prompt = 'Prompt-1'
sorted_df = df_results_final.sort_values(prompt, ascending=False)
plt.figure(figsize = (10, 8))
ax = sns.barplot(y=sorted_df.index, x=prompt, data=sorted_df, errorbar=None)

for i, v in enumerate(sorted_df[prompt]):
    ax.annotate(str(round(v, 3)), xy=(v, i), ha='left', va='center', fontsize=8)

ax.set_xlim(0, 0.85)

plt.title(f'Barplot of {prompt} Results (QWK Scores)')
plt.show()
```
