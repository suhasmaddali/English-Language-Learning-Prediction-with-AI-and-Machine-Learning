### Feature Engineering

```python
def text_length_features(text):
    words = word_tokenize(text)
    chars = len(text)
    avg_word_length = sum(len(word) for word in words) / len(words)
    sentences = sent_tokenize(text)
    avg_sentence_length = sum(len(sent_tokenize(sentence)) for sentence in sentences) / len(sentences)

    return len(words), chars, avg_word_length, avg_sentence_length
```

```python
def lexical_diversity(text):
    words = word_tokenize(text)
    unique_words = set(words)
    type_token_ratio = len(unique_words) / len(words)
    content_words = [word for word in words if word.lower() not in stopwords.words('english')]
    lexical_density = len(content_words) / len(words)

    return type_token_ratio, lexical_density
```

```python
def syntactic_features(text):
    words = nltk.pos_tag(word_tokenize(text))
    pos_freq = nltk.FreqDist(tag for (word, tag) in words)
    pos_ratios = {tag: count / len(words) for (tag, count) in pos_freq.items()}
    num_sentences = len(sent_tokenize(text))
    num_paragraphs = text.count('\n\n') + 1

    return pos_ratios, num_sentences, num_paragraphs
```

```python
def average_words_per_sentence(essay):
    sentences = re.split(r'[.!?]+', essay)
    sentence_lengths = [len(sentence.split()) for sentence in sentences if sentence]
    avg_words_per_sentence = sum(sentence_lengths) / len(sentence_lengths)
    return avg_words_per_sentence
```

```python
def text_preprocessing(text, english_words = words):

    # Get a list of all English words
    english_words = set(english_words.words())

    # Define a regular expression to match only alphanumeric characters
    regex = re.compile('[^a-zA-Z0-9\s]')

    # Remove all punctuation characters and convert to lowercase
    text = regex.sub('', text.lower())

    # Split the text into words
    words = text.split()

    # Filter out words that are not English words
    english_words_only = [word for word in words if word in english_words]

    # Join the English words into a string
    english_text = ' '.join(english_words_only)
    
    return english_text
```
